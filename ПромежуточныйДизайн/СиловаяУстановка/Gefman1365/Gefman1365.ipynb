## Подсчет входного сопротивления

```python
vin = rho * D**5 * omega7000 * np.polyval(p7000cq, AR(7000, 20)) * 0.082 / (4*np.pi**2 * 60 / (2*np.pi * 900)) + 1.2*0.082 + 60 / (2*np.pi * 900) * omega7000
v = sp.Symbol('v', real = True, positive = True)

a = (rho * D**5 * float(p7000cq[-1])) / (4*np.pi**2)
b = (rho * D**4 * float(p7000cq[1])* 17)/(2*np.pi) + (60 / (2*np.pi * 900))**2 / 0.082 
c = rho * D**3 * 17**2 * float(p7000cq[0]) + 60 / (2*np.pi * 900) * (1.2 - v/0.082)

mega = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
sp.solve(mega - omega7000, v)
```
